```python
def funct(xxxx):
  return xxxx**2 + 3*xxxx -5
  #return x**2 + 3*x -5

xxxx = np.random.rand(201)
yyyy = funct(xxxx)
zzzz = np.random.randn(201)
kkkk = yyyy + zzzz

plt.plot(xxxx,kkkk,'.', label='data')
plt.plot(xxxx,yyyy, label='ground truth')
```

```python
m = np.linspace(-2,2,201)
n = funct(m)
k = np.random.normal(size=201)
e = n + k

plt.plot(m,e,'.', label='data')
plt.plot(m,n, label='ground truth')

# function to show the plot
plt.show()
```

```python
import tensorflow as tf

class NewModel(tf.Module):
  def __init__(self):
    randinit = tf.random.uniform(shape=[3], minval=0., maxval=5., seed=22)
    print("randinit ",randinit)
    self.w_1 = tf.Variable(randinit[0])
    self.w_2 = tf.Variable(randinit[1])
    self.b = tf.Variable(randinit[2])


  @tf.function
  def __call__(self, x):
    return self.w_1 * (x**2) + self.w_2 * x +  self.b
```

```python
m_t = tf.convert_to_tensor(m, dtype=tf.float32)
n_t = tf.convert_to_tensor(n, dtype=tf.float32)
p = NewModel()
```

```python
m_t
```

```python
plt.plot(m,e,'.', label='data')
plt.plot(m,n, label='ground truth')
plt.plot(m,p(m_t), label='Prediction')
plt.title('Before training')
plt.legend()
# function to show the plot
plt.show()
```

```python
def mse_loss(y, y_pred):
  return tf.reduce_mean(tf.square(y_pred-y))
```

```python
batch_size=32
dataset = tf.data.Dataset.from_tensor_slices((m_t,n_t))
dataset = dataset.shuffle(buffer_size=m_t.shape[0]).batch(batch_size)
```

```python
dataset
```

```python
epochs = 70
learning_rate = 0.01
losses = []
```

```python
for epoch in range(epochs):
  for x_bacth, y_batch in dataset:
    with tf.GradientTape() as tape:
      batch_loss = mse_loss(y_batch,p(x_bacth))
    grads = tape.gradient(batch_loss,p.variables)
    for g, v in zip(grads,p.variables):
      v.assign_sub(learning_rate*g)

  loss_e = mse_loss(n_t,p(m_t))
  losses.append(loss_e)
  if epoch % 10 == 0:
    print(f'Mean squared error for step {epoch}: {loss_e.numpy():0.3f}')

# Plot model results
print("\n")
plt.plot(range(epochs), losses)
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error (MSE)")
plt.title('MSE loss vs training iterations');
```

```python
plt.plot(m,e,'.', label='data')
plt.plot(m,n, label='ground truth')
plt.plot(m,p(m_t), label='Prediction')
plt.title('After training')
plt.legend()
# function to show the plot
plt.show()
```

```python
new_model = tf.keras.Sequential([
    tf.keras.layers.Lambda(lambda m_t: tf.stack([m_t, m_t**2], axis=1)),
    tf.keras.layers.Dense(units=1, kernel_initializer=tf.random.normal)])
```

```python
new_model.compile(
    loss=tf.keras.losses.MSE,
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01))

history = new_model.fit(m_t, n_t,
                        epochs=70,
                        batch_size=32,
                        verbose=0)
```

```python
plt.plot(history.history['loss'])
plt.xlabel('Epoch')
plt.ylim([0, max(plt.ylim())])
plt.ylabel('Loss [Mean Squared Error]')
plt.title('Keras training progress');
```

```python
plt.plot(m,e,'.', label='data')
plt.plot(m,n, label='ground truth')
plt.plot(m,new_model(m_t), label='Prediction')
plt.title('After Keras training')
plt.legend()
# function to show the plot
plt.show()
```
